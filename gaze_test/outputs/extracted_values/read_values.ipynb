```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_names = [
    "index_4.txt",
    "index_33.txt",
    "index_45.txt",
    "index_48.txt"
]
```

```python
data = {}
for file_name in file_names:
    file_path = os.path.join(file_name)
    if os.path.exists(file_path):
        with open(file_path, "r") as file:
            values = [float(line.strip()) for line in file.readlines()]
            data[file_name] = values
    else:
        print(f"File not found: {file_path}")
```

```python
# Display the data for each file
for file_name, values in data.items():
    print(f"Data from {file_name}:")
    print(values[:10], "...")  # Print first 10 values for brevity

# Optionally visualize the data
plt.figure(figsize=(10, 6))
for file_name, values in data.items():
    plt.plot(values, label=file_name)

plt.title("Values from Index Files")
plt.xlabel("Index")
plt.ylabel("Value")
plt.legend()
plt.grid()
plt.show()
```

```python
sum(data)
```

```python
indices = [4, 33, 45, 48]

# Initialize a dictionary to store the sums
sums = {str(index): 0 for index in indices}

# Iterate through the data to compute the sums
for file_name, values in data.items():
    index = file_name.split("_")[1].split(".")[0]  # Extract index from file name (e.g., "index_4.txt" -> "4")
    if index in sums:
        sums[index] += sum(values)

# Print the results
for index, total in sums.items():
    print(f"Sum for index {index}: {total}")
```
